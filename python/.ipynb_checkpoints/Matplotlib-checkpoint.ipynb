```python
figure = plt.figure()
axes = figure.add_subplot()

x = [1,2,3,4]
y = [2,4,6,8]
x2 = [1,2,3,4]
y2 = [-2,-4,-6,-8]
axes.plot(x,y, marker='o')
axes.plot(x2,y2)
plt.show()
```
